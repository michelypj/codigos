```python
filist = np.loadtxt('fileList.dat', str) 

#print(filist)
```

```python
print(filist[400],filist[-1])
```

```python
rte=[]
for i in filist[:-5]:
    fn = i
    fn = fn.replace('b','')
    fn = fn.replace('\'','')    
    outfile = loadtxt(fn, float)
    l=sum(outfile[:,2])/len(outfile[:,2])
    rte.append(l)
    #print(sum(outfile[:,2])/len(outfile[:,2]))
    #print(sum(outfile[:,6])/len(outfile[:,6]))
```

```python
rt=[]
for i in filist[:-5]:
    fn = i
    fn = fn.replace('b','')
    fn = fn.replace('\'','')    
    outfile = loadtxt(fn, float)
    rt.append(sum(outfile[:,6])/len(outfile[:,6]))
    #print(sum(outfile[:,2])/len(outfile[:,2]))
    #print(sum(outfile[:,6])/len(outfile[:,6]))
```

```python
#print(rte)
print(len(rt),len(rte))
```

```python
plt.plot(rte,rt,'o')
plt.grid()
plt.show()
```

```python
from scipy.optimize import curve_fit
```

```python
def lineal(x, m, b):
    return m*x + b 
pars, cov = curve_fit(f=lineal, xdata=rte, ydata=rt, p0=[0, 0])
```

```python
print(pars)
```

```python
plt.plot(rte,rt,'o')
tmp=np.array(rte)
plt.plot(tmp,lineal(tmp,*pars),'-')
plt.grid()
plt.show()
```

```python
p0=920
s0=305
flx_corr=pars[0]*(tmp-p0)+s0
```
